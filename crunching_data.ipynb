# Ticket data

### Descrition des features:


*   ***id***: Id du ticket
*   ***company***: Companie de voyage (compagnie de train par exemple)
*   ***o_station, d_station***: Station de départ (**O**rigin station) et d'arrivée (**D**estination station)
*   ***departure_ts, arrival_ts***: Heure de départ et heure d'arrivée
*   ***price_in_cents***: Prix du ticket
*   ***search_ts***: L'heure de la recherche des voyages
*   ***middle_stations***: Les arrêts (stations) de train ou du bus entre la station du départ et celle d'arrivée
*   ***other_companies***: autres companies qui propose le voyage
*   ***o_city, d_city***: Ville de départ (**O**rigin ville) et d'arrivée (**D**estination station)

* Les variables o_station et d_station sont des variables qualitatives. On doit changer leur type.

## Relation entre durée du voyage et prix du ticket

Pour faire cela calculons une variable durée (en heure) de voyage à partir des deux variables heure de départ et heure d'arrivée

```python
# Calculer la nouvelle variable 
data_ticket['duration'] = (data_ticket['arrival_ts'] - data_ticket['departure_ts'])/ pd.Timedelta(1, unit='h')
```

```python
data_ticket
```

#### Relation entre la durée et le prix:

```python
def whiskers(df, column):
    # calculate the quartiles
    q1 = np.percentile(df[column], 25)
    q3 = np.percentile(df[column], 75)

    # calculate the interquartile range (IQR)
    iqr = q3 - q1

    # calculate the lower and upper whiskers
    lower_whisker = q1 - 1.5 * iqr
    upper_whisker = q3 + 1.5 * iqr

    # print the results
    print('Lower whisker for',column, ':', lower_whisker)
    print('Upper whisker for',column, ':', upper_whisker)
```

```python
fig, ax = plt.subplots(figsize=(6, 8))
plt.boxplot(x=data_ticket['duration'])

plt.xlabel('Variable')
plt.ylabel('Value')
plt.title('Boxplot of variable')

plt.show()
```

```python
whiskers(data_ticket,'duration')
```

```python
fig, ax = plt.subplots(figsize=(6, 8))
plt.boxplot(x=data_ticket['price_in_cents'])

plt.xlabel('Variable')
plt.ylabel('Value')
plt.title('Boxplot of variable')

plt.show()
```

```python
whiskers(data_ticket,'price_in_cents')
```

```python
data_ticket['price_in_cents'].min()
```

```python
less = data_ticket[data_ticket['duration'] <= 28]
less = less[less['price_in_cents'] <= 17025]
less
```

```python
sample_df = less.sample(frac=0.1, random_state=42)
sample_df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.set(style='ticks', color_codes=True)
sns.regplot(x='duration', y='price_in_cents', data=sample_df, ax=ax, line_kws={"color": "r", "alpha": 0.7, "lw": 2})
corr_coef = sample_df['duration'].corr(sample_df['price_in_cents'])
plt.title(f"Scatter plot of x and y (correlation = {corr_coef:.2f})")
plt.show()
```
