# Ticket data

### Descrition des features:


*   ***id***: Id du ticket
*   ***company***: Companie de voyage (compagnie de train par exemple)
*   ***o_station, d_station***: Station de départ (**O**rigin station) et d'arrivée (**D**estination station)
*   ***departure_ts, arrival_ts***: Heure de départ et heure d'arrivée
*   ***price_in_cents***: Prix du ticket
*   ***search_ts***: L'heure de la recherche des voyages
*   ***middle_stations***: Les arrêts (stations) de train ou du bus entre la station du départ et celle d'arrivée
*   ***other_companies***: autres companies qui propose le voyage
*   ***o_city, d_city***: Ville de départ (**O**rigin ville) et d'arrivée (**D**estination station)

## Nettoyage des données

```python
data_ticket.dtypes
```

Les variables o_station et d_station sont des variables qualitatives. On doit changer leur type.

```python
data_ticket.o_station = data_ticket.o_station.astype(int)
data_ticket.d_station = data_ticket.d_station.astype(int)
data_ticket.dtypes
```

## Distribution de companies

```python
# get the count of unique values in companies
companies = data_ticket.company.value_counts()


companies.plot(kind='bar')
plt.xlabel('company')
plt.ylabel('Count')
plt.show()
```

```python
companies = data_ticket.company.value_counts()/data_ticket.shape[0] *100


companies.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
# get the top 5 values of 'd_city' column based on value counts
top_10_o_city = data_ticket['o_city'].value_counts().nlargest(3).index.tolist()
top_10_d_city = data_ticket['d_city'].value_counts().nlargest(3).index.tolist()

# filter the dataframe based on the values in the top 5 list
df_top_10 = data_ticket[data_ticket['o_city'].isin(top_10_o_city) | data_ticket['d_city'].isin(top_10_d_city)]


# display the top 5 rows of the filtered dataframe
df_top_10
```

```python
ct = pd.crosstab(data_ticket['o_city'], data_ticket['d_city'])

# plot the contingency table as a stacked bar chart
ct.plot(kind='bar', stacked=True)

# add labels and title
plt.xlabel('origin ciy')
plt.ylabel('Count')
plt.title('Profile Chart ')

# display the chart
plt.show()
```
